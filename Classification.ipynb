# 1. Impoting Libraries

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing Dataset

```python
fashion_mnist = tf.keras.datasets.fashion_mnist

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
X_test.shape
```

# 2. Data Anlysis and Visualisation

```python
X_train.shape
```

```python
for i in range(9):
    plt.imshow(X_train[i], cmap=plt.cm.binary)
    plt.xlabel(class_names[y_train[i]])
    plt.show()
```

# 3. Classification with Deep Learning Algorithms

```python
X_train, X_test = X_train/255, X_test/255
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10)
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, epochs=10)
```

```python
probability_model = tf.keras.Sequential([model, 
                                         tf.keras.layers.Softmax()])
```

```python
train_predictions = probability_model.predict(X_train)
test_predictions = probability_model.predict(X_test)
```

```python
y_train_pred = np.array([np.argmax(train_predictions[i]) for i in range(len(y_train))])
y_test_pred = np.array([np.argmax(test_predictions[i]) for i in range(len(y_test))])
```

```python
from sklearn.metrics import accuracy_score

train_score = accuracy_score(y_train_pred, y_train)
test_score = accuracy_score(y_test_pred, y_test)
```

```python
print("train_score=",train_score*100,"%")
```

```python
print("test_score=",test_score*100,"%")
```

# 4. Classification with Machine Learning Algorithms

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
X_train = X_train.reshape(60000,28*28)
X_test = X_test.reshape(10000,28*28)
```

```python
models = {}


models["Logistic Regression"] = LogisticRegression(multi_class="ovr")
models["Random Forest Classification"] = RandomForestClassifier(n_estimators=100)
models["KNearest Classification"] = KNeighborsClassifier()
models
```

```python
test_scores = []
train_scores = []
for name, model in models.items():
    model.fit(X_train, y_train)
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    train_score = accuracy_score(y_train_pred, y_train)
    test_score = accuracy_score(y_test_pred, y_test)
    
    test_scores.append(name+" :"+str(test_score*100)+"%")
    train_scores.append(name+" :"+str(train_score*100)+"%")
    
test_scores
```
